```python
import matplotlib.pyplot as plt
import numpy as np

# definng the time range 
t = np.linspace(0.1, 10, 100)  

# defining the scale factor a(t) as a constant times t
a = np.sqrt(np.abs(-1)) * t  # assuming |k| = 1 for simplicity

# creating the plot 
plt.figure(figsize=(8, 6))
plt.plot(t, a, label=r'$a(t) = \sqrt{|k|} \, t$', color='magenta')

# annotating the domination part again
plt.text(5, a[50], 'is the domination part', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# labels & titles 
plt.xlabel('Time $t$', fontsize=14)
plt.ylabel('Scale factor $a(t)$', fontsize=14)
plt.title('Graph of the Scale Factor $a(t)$ over Time with Constant Expansion', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.savefig('plot.png', dpi=700)
plt.show()

```
